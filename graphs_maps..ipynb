```python
import folium
from folium import IFrame 
from branca.element import Template, MacroElement
import geopandas as gpd
```

```python
gdf = gpd.read_file('ghmc-area.geojson') 
```

```python
gdf1 = gdf.drop(columns=['@timestamp'])
```

```python
center = gdf1.geometry.centroid.y.mean(), gdf1.geometry.centroid.x.mean()
```

```python
area_map = folium.Map(location=center, zoom_start=12)

# Add the GeoJSON data from the updated gdf1 to the map
folium.GeoJson(gdf1).add_to(area_map)

# Extract the first element (area names) from top_list and bottom_list
top_areas = [item[0] for item in price_top_list]
# bottom_areas = [item[0] for item in bottom_5_list]

# Create a FeatureGroup for the markers
fg = folium.FeatureGroup(name='map1')
```

```python
for i in a_list:
    popup_html = f"<b style='font-size:15px'>{i[0]},{i[1]}</b>"
    iframe = folium.IFrame(html=popup_html, width=150, height=70)
    if i[0] in top_areas:
        fg.add_child(folium.Marker(location=[i[2], i[3]], popup=folium.Popup(iframe), 
                                   icon=folium.Icon(color='red', icon_size=(40, 40))))
    # elif i[0] in bottom_areas:
    #     fg.add_child(folium.Marker(location=[i[2], i[3]], popup=folium.Popup(iframe), 
    #                                icon=folium.Icon(color='blue', icon_size=(40, 40))))
    else:
        fg.add_child(folium.Marker(location=[i[2], i[3]], popup=folium.Popup(iframe), 
                                   icon=folium.Icon(color='green', icon_size=(40, 40))))

# Add FeatureGroup with the markers to the map
area_map.add_child(fg)

legend_html = '''
<div style="
    position: fixed; 
    bottom: 50px; left: 50px; width: 150px; height: 120px; 
    background-color: white; 
    border:2px solid grey; z-index:9999; font-size:14px;
    padding: 10px;
    ">
    <b>Legend</b><br>
    <i class="fa fa-map-marker fa-2x" style="color:red"></i>&nbsp; Top Areas <br>
    <i class="fa fa-map-marker fa-2x" style="color:green"></i>&nbsp; Other Areas <br>
</div>
'''

area_map.get_root().html.add_child(folium.Element(legend_html))

area_map.save('F.html')

```
